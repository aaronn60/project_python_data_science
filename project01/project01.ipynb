```python
import pandas as pd
```

```python
df= pd.read_csv("train.csv")
df.head()
```

```python
# cek jumlah baris & kolom
df.shape
# cek tipe data tiap kolom
df.info()
# cek statistik tiap kolom
df.describe() 
```

```python
#check missing value tiap kolom
df.isnull().sum()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#analisis total survived
ax= sns.countplot(x="Survived", data=df)
ax.bar_label(ax.containers[0])
plt.show()
```

```python
#analisis histogram tabel age dengan drop missing value di kolom age, dibagi ke 20 bins
sns.histplot(df["Age"].dropna(), bins =20)
plt.show()
```

```python
#menambahkan kurva distribusi data (Kernel Density Estimate=KDE) pada histplot
sns.histplot(df["Age"].dropna(), bins =20, kde=True)
plt.show()
```

```python
#distribusi Gender
ax= sns.countplot (x="Sex",data=df)
ax.bar_label(ax.containers[0])
plt.show()
```

```python
#komposisi male lebih mendominasi dibandingkan female
```

```python
#analisis total survived berdasarkan gender
ax= sns.countplot(x="Sex", hue = "Survived", data=df)
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.show()
```

```python
#female lebih banyak yang survive dibandingkan male
```

```python
#analisis total survived berdasarkan kelas
ax = sns.countplot(x="Pclass", hue="Survived", data=df)
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.show()
```

```python
#Penumpang kelas 1 memiliki survival rate yang lebih tinggi dibanding kelas lainnya
#Penumpang kelas 3 memiliki survival rate paling rendah dibandingkan kelas lainnya
```
